# Basic Usage

## 1. Code Completion

Copilot provides suggestions as you type. Let's try some examples:

1. Start typing a function definition:
```python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
```

2. Copilot will suggest the implementation. Press Tab to accept or keep typing for different suggestions.

````python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
    if height <= 0:
        raise ValueError("Height must be greater than zero.")
    if weight <= 0:
        raise ValueError("Weight must be greater than zero.")
    bmi = weight / (height ** 2)
    return round(bmi, 2)

def bmi_category(bmi):
    # Determine BMI category based on the calculated BMI
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obesity"
def main():
    # Example usage
    try:
        weight = float(input("Enter your weight in kg: "))
        height = float(input("Enter your height in meters: "))
        bmi = calculate_bmi(weight, height)
        category = bmi_category(bmi)
        print(f"Your BMI is {bmi} and you are classified as {category}.")
    except ValueError as e:
        print(e)
if __name__ == "__main__":
    main()
# This code calculates BMI and categorizes it
# It includes error handling for invalid inputs
# The code is structured to be reusable and maintainable
# The functions are defined separately for clarity
# The main function handles user input and output
# The code is designed to be user-friendly and informative  
    
````

## 2. Comment-to-Code

One of Copilot's most powerful features is generating code from comments:


````python
# Create a function that converts temperature from Celsius to Fahrenheit
# Include input validation and round to 2 decimal places



# Try typing this comment and let Copilot suggest the implementation

def celsius_to_fahrenheit(celsius):
    """
    Convert temperature from Celsius to Fahrenheit
    Args:
        celsius (float): Temperature in Celsius
    Returns:
        float: Temperature in Fahrenheit, rounded to 2 decimal places
    """
    # Input validation
    if not isinstance(celsius, (int, float)):
        raise TypeError("Temperature must be a number")
    
    # Convert to Fahrenheit using the formula: °F = (°C × 9/5) + 32
    fahrenheit = (celsius * 9/5) + 32
    return round(fahrenheit, 2)

# Example usage
def main():
    try:
        temp = float(input("Enter temperature in Celsius: "))
        result = celsius_to_fahrenheit(temp)
        print(f"{temp}°C is equal to {result}°F")
    except ValueError:
        print("Please enter a valid number")
    except TypeError as e:
        print(e)

if __name__ == "__main__":
    main()
````

## 3. Working with Libraries

Copilot is particularly helpful when working with libraries. It can suggest proper import statements and usage patterns:

````python
# Create a scatter plot using matplotlib with the following features:
# - Random data points
# - Custom colors based on point values
# - A trend line
# - Proper labels and title
# - A legend
# - Proper

# Let Copilot suggest the implementation
import numpy as np
import matplotlib.pyplot as plt

# Generate random data points
np.random.seed(42)
x = np.random.rand(50) * 100  # Random x values between 0 and 100
y = 2 * x + np.random.randn(50) * 10  # y = 2x + noise

# Custom colors based on y values
colors = np.where(y > np.mean(y), 'blue', 'orange')

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=colors, label='Data Points', alpha=0.7)

# Add a trend line
m, b = np.polyfit(x, y, 1)  # Linear regression
plt.plot(x, m * x + b, color='red', label='Trend Line')

# Add labels, title, and legend
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot with Trend Line and Custom Colors')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
````

### ✏️ Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
class PatientRecords:
    def __init__(self):
        """Initialize an empty dictionary to store patient records."""
        self.records = {}

    def add_record(self, patient_id, name, age, diagnosis):
        """
        Add a new patient record.
        Args:
            patient_id (str): Unique identifier for the patient.
            name (str): Name of the patient.
            age (int): Age of the patient.
            diagnosis (str): Diagnosis of the patient.
        """
        if patient_id in self.records:
            raise ValueError("Patient ID already exists.")
        if not isinstance(age, int) or age <= 0:
            raise ValueError("Age must be a positive integer.")
        self.records[patient_id] = {"name": name, "age": age, "diagnosis": diagnosis}

    def update_record(self, patient_id, **kwargs):
        """
        Update an existing patient record.
        Args:
            patient_id (str): Unique identifier for the patient.
            kwargs: Fields to update (e.g., name, age, diagnosis).
        """
        if patient_id not in self.records:
            raise ValueError("Patient ID not found.")
        for key, value in kwargs.items():
            if key == "age" and (not isinstance(value, int) or value <= 0):
                raise ValueError("Age must be a positive integer.")
            if key in self.records[patient_id]:
                self.records[patient_id][key] = value

    def get_record(self, patient_id):
        """
        Retrieve a patient record.
        Args:
            patient_id (str): Unique identifier for the patient.
        Returns:
            dict: Patient record.
        """
        if patient_id not in self.records:
            raise ValueError("Patient ID not found.")
        return self.records[patient_id]

    def average_age(self):
        """
        Calculate the average age of all patients.
        Returns:
            float: Average age.
        """
        if not self.records:
            return 0
        total_age = sum(record["age"] for record in self.records.values())
        return total_age / len(self.records)

    def count_by_diagnosis(self, diagnosis):
        """
        Count the number of patients with a specific diagnosis.
        Args:
            diagnosis (str): Diagnosis to count.
        Returns:
            int: Number of patients with the diagnosis.
        """
        return sum(1 for record in self.records.values() if record["diagnosis"] == diagnosis)
````

### ✏️ Exercise: Create a data visualization dashboard
   - Use pandas for data manipulation
   - Create multiple types of plots using matplotlib
   - Add interactive elements

````python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

# Sample data
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, 35, 40, 45],
    "Score": [88, 92, 85, 95, 90],
    "Category": ["A", "B", "A", "B", "A"]
}
df = pd.DataFrame(data)

# Function to display a bar plot
def plot_bar(column):
    plt.figure(figsize=(8, 5))
    df[column].value_counts().plot(kind="bar", color="skyblue")
    plt.title(f"Bar Plot of {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.show()

# Function to display a scatter plot
def plot_scatter(x_column, y_column):
    plt.figure(figsize=(8, 5))
    plt.scatter(df[x_column], df[y_column], color="green")
    plt.title(f"Scatter Plot: {x_column} vs {y_column}")
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()

# Interactive dashboard
def dashboard(plot_type, column=None, x_column=None, y_column=None):
    if plot_type == "Bar Plot" and column:
        plot_bar(column)
    elif plot_type == "Scatter Plot" and x_column and y_column:
        plot_scatter(x_column, y_column)
    else:
        print("Please select valid options.")

# Widgets for interactivity
plot_type_widget = widgets.Dropdown(
    options=["Bar Plot", "Scatter Plot"],
    description="Plot Type:"
)
column_widget = widgets.Dropdown(
    options=df.columns,
    description="Column:"
)
x_column_widget = widgets.Dropdown(
    options=df.columns,
    description="X Column:"
)
y_column_widget = widgets.Dropdown(
    options=df.columns,
    description="Y Column:"
)

# Interactive function
interact(
    dashboard,
    plot_type=plot_type_widget,
    column=column_widget,
    x_column=x_column_widget,
    y_column=y_column_widget
)
````
